```python
import pandas as pd
import matplotlib.pyplot as plt



plt.style.use("seaborn")

file = pd.read_excel("coordinate.xlsx")

x_Easom_FA= file["x_Easom_FA"]
y_Easom_FA= file["y_Easom_FA"]
x_Easom_GA= file["x_Easom_GA"]
y_Easom_GA= file["y_Easom_GA"]

x_Dropwave_FA = file["x_Dropwave_FA"]
y_Dropwave_FA = file["y_Dropwave_FA"]
x_Dropwave_GA = file["x_Dropwave_GA"]
y_Dropwave_GA = file["y_Dropwave_GA"]

x_Shubert_FA = file["x_Shubert_FA"]
y_Shubert_FA = file["y_Shubert_FA"]
x_Shubert_GA = file["x_Shubert_GA"]
y_Shubert_GA = file["y_Shubert_GA"]


x_Sphere_FA_2 = file["x_Sphere_FA_2"]
y_Sphere_FA_2 = file["y_Sphere_FA_2"]
x_Sphere_GA_2 = file["x_Sphere_GA_2"]
y_Sphere_GA_2 = file["y_Sphere_GA_2"]


x_Rosenbrock_FA_2 = file["x_Rosenbrock_FA_2"]
y_Rosenbrock_FA_2 = file["y_Rosenbrock_FA_2"]
x_Rosenbrock_GA_2 = file["x_Rosenbrock_GA_2"]
y_Rosenbrock_GA_2 = file["y_Rosenbrock_GA_2"]



plt.scatter(x_Easom_FA,y_Easom_FA, s=100, c="red", edgecolor="black", linewidth=1, alpha=0.75, label="FA")
plt.scatter(x_Easom_GA,y_Easom_GA, s=100, c="blue", edgecolor="black", linewidth=1, alpha=0.75, label="GA")
plt.legend(loc='best')
plt.title("FUNZIONE DI EASOM")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.scatter(x_Dropwave_FA,y_Dropwave_FA, s=100, c="red", edgecolor="black", linewidth=1, alpha=0.75, label="FA")
plt.scatter(x_Dropwave_GA,y_Dropwave_GA, s=100, c="blue", edgecolor="black", linewidth=1, alpha=0.75, label="GA")
plt.legend(loc='best')
plt.title("FUNZIONE DROPWAVE")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


plt.scatter(x_Shubert_FA,y_Shubert_FA, s=100, c="red", edgecolor="red", linewidth=5, alpha=0.75, label="FA")
plt.scatter(x_Shubert_GA,y_Shubert_GA, s=100, c="blue", edgecolor="black", linewidth=1, alpha=0.75, label="GA")
plt.legend(loc='best')
plt.title("FUNZIONE DI SHUBERT")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


plt.scatter(x_Sphere_FA_2,y_Sphere_FA_2, s=100, c="red", edgecolor="black", linewidth=1, alpha=1, label="FA")
plt.scatter(x_Sphere_GA_2,y_Sphere_GA_2, s=100, c="blue", edgecolor="black", linewidth=1, alpha=0.75, label="GA")
plt.legend(loc='best')
plt.title("FUNZIONE DELLA SFERA (DIM=2)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


plt.scatter(x_Rosenbrock_FA_2,y_Rosenbrock_FA_2, s=100, c="red", edgecolor="red", linewidth=5, alpha=1, label="FA")
plt.scatter(x_Rosenbrock_GA_2,y_Rosenbrock_GA_2, s=100, c="blue", edgecolor="black", linewidth=1, alpha=0.75, label="GA")
plt.legend(loc='best')
plt.title("FUNZIONE DI ROSENBROCK (DIM=2)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
